```python
import pandas as pd
import numpy as np
from sklearn.ensemble import ExtraTreesClassifier

TrainData = pd.read_csv("./LargeTrain.csv")
ClassData = pd.read_csv("../trainLabels.csv")
X = TrainData.values[:,0:1805]
Y = ClassData.values[:,1].astype(float)

# feature extraction
forest = ExtraTreesClassifier(n_estimators=100, random_state=0)
forest.fit(X, Y)

importances = forest.feature_importances_
indices = np.argsort(importances)[::-1] # from largest to smallest
attribute = TrainData.columns

# Print the feature ranking
print("Feature ranking:")
for f in range(X.shape[1]):
    print("%d. feature %d: %s ; ranking: %f" % (f + 1, indices[f] + 1, attribute[indices[f]], importances[indices[f]])) # show the result index from 1, not 0
```
